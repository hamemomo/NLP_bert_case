```python
import torch
from transformers import BertTokenizer, BertModel, BertForMaskedLM,BertConfig

def encode(text):
    PRETRAINED_MODEL_NAME = "bert-base-chinese"
    
    tokenizer = BertTokenizer.from_pretrained(PRETRAINED_MODEL_NAME)
    
    
    tokenized_text = ["[CLS]"]
    tokens_a = tokenizer.tokenize(text)
    tokenized_text += tokens_a + ["[SEP]"]
    
    #configuration = BertConfig(hidden_size=768)

# Initializing a model from the bert-base-uncased style configuration
    #model = BertModel(configuration)
    
    
    
    
    #tokenized_text = tokenizer.tokenize(text) #分字
    indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text) #字轉換成 ID
    #print(tokenized_text)

    segments_ids = [1] * len(tokenized_text)
    tokens_tensor = torch.tensor([indexed_tokens])
    segments_tensors = torch.tensor([segments_ids])
    model = BertModel.from_pretrained(PRETRAINED_MODEL_NAME)

    tokens_tensor = tokens_tensor.cuda()
    segments_tensors = segments_tensors.cuda()
    model.cuda()
    configuration = model.config
    print(configuration)
    encoded_layers, pooled_output= model(tokens_tensor, segments_tensors)
    
    print(encoded_layers.shape)
    return pooled_output

```
