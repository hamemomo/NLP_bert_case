```python

import requests
from bs4 import BeautifulSoup
#import time
article_href = []
url_1 = 'https://www.ptt.cc/'
#url_2 = 'bbs/forsale/index'+str(1885)+'.html'
#url = url_1+url_2
num = 1885
#print(url)
for i in range(1885,1000,-1):
    print('i :',i)
    url_2 = 'bbs/forsale/index'+str(i)+'.html'
    url = url_1+url_2
    print(url)
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"html.parser")
    results = soup.select("div.title")
    #results[1].select('a').get('href')
    for item in results:
        try : 
            item_href = item.select("a")[0].get('href')
            print(item_href)
        except:
            item_href = '本文已刪除'
            #print("本文已刪除")

        article_href.append(item_href)
    nuum = num-1
    #time.sleep(0)
    print(item_href)
```

```python
for i in article_href:
    if i == '本文已刪除':
        article_href.remove('本文已刪除')
```

```python
print(len(article_href))
```

```python
product_title_full = [None] * len(article_href)
product_time = [None] * len(article_href)
product_title = [None] * len(article_href)
product_price = [None] * len(article_href)
product_source= [None] * len(article_href)
product_sell_price =[None] * len(article_href)
product_state =[None] * len(article_href)

for index,url_product in enumerate(article_href):
    
    print(index)
    if url_product =='本文已刪除':
        print('本文已刪除')
        print('========================================================')
    else:
        url = url_1+url_product
        r = requests.get(url)
        soup = BeautifulSoup(r.text,"html.parser")
        results = soup.select('span.article-meta-value')
        #print(len(results))
        if len(results)>3:
            #作者 看板 標題 時間
            print('作者 : ',results[0].text)
            
            print('看板 : ',results[1].text)
            print('標題 : ',results[2].text)
            print('時間 : ',results[3].text)
            product_title_full[index] = results[2].text
            product_time[index] = results[3].text
            #content 文章內文
            content = soup.find(id="main-content").text
            
            target_content = u'※ 發信站: 批踢踢實業坊(ptt.cc),'
            #print('完整',content)
            #去除掉 target_content
            content = content.split(target_content)
            content_row = content[0].split('\n')
          
            for i in range(len(content_row)):
                if '商品名稱' in content_row[i]:   
                    print(content_row[i][5:])
                    product_title[index] = content_row[i][5:]
                    
                    
                elif '商品來源' in content_row[i]:
                    print(content_row[i][5:])
                    product_source[index] = content_row[i][5:]
                    
                elif '商品價格' in content_row[i]:
                    print(content_row[i][5:])
                    product_price[index] = content_row[i][5:]
                    
                    
                elif '出售金額' in content_row[i]:
                    print(content_row[i][5:])
                    product_sell_price[index] = content_row[i][5:]
                    
                elif '新舊狀況' in content_row[i]:
                    print(content_row[i][5:])
                    product_state[index] = content_row[i][5:]
                    
                    
                    
                    
                    
                    
                    
                    
                    
                    print('---產品分隔---')
            print('========================================================')
        else:
            continue;
        
```

```python
import pandas as pd
df = pd.DataFrame(
    list(zip(product_title_full,product_time,product_title,product_price,product_source,product_sell_price,product_state)),
                  columns =['full_title','time','title', 'price','source','sell_price','state']) 
```

```python
df.to_csv('二手估價_test.csv',encoding='utf_8_sig')
```
