```python
import xmltodict
import itertools
import numpy as np

# read the sample file and parse to dict type
sample_xml = "articles.xml"
with open(sample_xml, encoding='utf-8') as fd:
    sample = xmltodict.parse(fd.read())
```

```python

# split the data to article list
articles = dict(sample['articles'])['article']

# get the list of title from article list
titles = [dict(article)['title'] for article in articles]
```
